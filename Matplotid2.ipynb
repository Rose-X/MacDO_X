[Exercise 1: Simple Polynomial Plot]

```python
import numpy as np
import matplotlib.pyplot as plt

#on génère un tableau avec des nb alt
x = np.linspace(-10, 10, 100)
#print(x)


# Définition du polynôme y = 2x^3 - 5x^2 + 3x - 7
y = 2*x**3 - 5*x**2 + 3*x - 7
#print(y)


# Configuration de la taille du  graphique
plt.figure(figsize=(10, 6))

# On trace la courbe
plt.plot(x, y, 'b-', label=r'$y = 2x^3 - 5x^2 + 3x - 7$')


plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomial Function Plot")
plt.legend()
plt.grid(True)

# Affichage du graphique
plt.show()
```

### Exercise 2: Exponential and Logarithmic Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 10, 500)

y1 = np.exp(x)  # On cal la fonction exponentielle e^x
y2 = np.log(x)  # idem pour naturel ln(x)

# Création de la figure
plt.figure(figsize=(10, 6))

# Tracé des fonctions avec des couleurs et styles différents
plt.plot(x, y1, 'r-', label=r'$y = e^x$')   # Rouge, ligne continue
plt.plot(x, y2, 'b--', label=r'$y = \log(x)$')  # Bleu, ligne pointillée

plt.xlabel("x")
plt.ylabel("y")
plt.title("Exponential and Logarithmic Functions")
plt.legend()
plt.grid(True)

plt.savefig("exponential_logarithmic_plot.png", dpi=100)

plt.show()
```

## Exercise 3: Figure and Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-2 * np.pi, 2 * np.pi, 500)  # Pour tan(x) et arctan(x)
x2 = np.linspace(-2, 2, 500)  # Pour sinh(x) et cosh(x)

y1 = np.tan(x1)
y2 = np.arctan(x1)
y3 = np.sinh(x2)
y4 = np.cosh(x2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 1 graph
ax1.plot(x1, y1, 'b-', label='tan(x)')  # Bleu continu
ax1.plot(x1, y2, 'r--', label='arctan(x)')  # Rouge pointillé
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('tan(x) and arctan(x)')
ax1.legend()
ax1.grid(True)
ax1.set_ylim(-10, 10)  #

# 2 graph : sinh(x) et cosh(x)**
ax2.plot(x2, y3, 'g-', label='sinh(x)')  # Vert continu
ax2.plot(x2, y4, 'm--', label='cosh(x)')  # Magenta pointillé
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('sinh(x) and cosh(x)')
ax2.legend()
ax2.grid(True)


plt.tight_layout()
plt.show()
```

### Exercise 6: Contour Plot

### Exercise 15: 3D Vector Field

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #Obligatoire
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
z = np.linspace(-5, 5, 10)
X, Y, Z = np.meshgrid(x, y, z)
U = -Y
V = X
W = Z

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, normalize=True)
ax.set_title('3D Vector Field')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

### Exercise 19: 3D Heatmap

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

heatmap = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

cbar = fig.colorbar(heatmap, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('Valeur de Z')

ax.set_xlabel('Axe X')
ax.set_ylabel('Axe Y')
ax.set_zlabel('Axe Z')
ax.set_title('Heatmap 3D de Z = sin(sqrt(x^2 + y^2))')

plt.show()

```

### **Exercise 20: 3D Density Plot**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Générer 1000 points aléatoires dans l'espace 3D
np.random.seed(42)
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
z = np.random.normal(0, 1, 1000)

# Calculer la densité des points
xyz = np.vstack([x, y, z])
density = gaussian_kde(xyz)(xyz)

idx = density.argsort()
x, y, z, density = x[idx], y[idx], z[idx], density[idx] # Créer un tracé 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(x, y, z, c=density, cmap='viridis', alpha=0.7) #trace

cbar = fig.colorbar(sc, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('Densité')

ax.set_xlabel('Axe X')
ax.set_ylabel('Axe Y')
ax.set_zlabel('Axe Z')
ax.set_title('Tracé de densité 3D des points aléatoires')

plt.show()
```
